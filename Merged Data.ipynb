```python
#import libraries
import pandas as pd 
import numpy as np
```

```python
#import csv files
customer = pd.read_csv(r"data\Customer Table.csv")
order = pd.read_csv(r"data\Order Table.csv")
product = pd.read_csv(r"data\Product Table.csv")
```

```python
#check for duplicates in the product table
product.duplicated().sum()
```

```python
#remove duplicates and reset the index
product = product.drop_duplicates().reset_index(drop = True)
```

```python
#check for duplicates in the order table
order.duplicated().sum()
```

```python
#check for duplicates in the customer table
customer.duplicated().sum()
```

```python
tables = [customer, order, product]

for item in tables:
    item.info()
```

```python
#the number countries with no territories assigned to them
customer[customer.TERRITORY.isna()].groupby('COUNTRY').count()
```

```python
#fill the empty territories with North America
customer.fillna('North America', inplace = True)
```

```python
#check  for null values
customer.isna().sum()
```

```python
#all territories mentioned
customer.TERRITORY.unique()
```

```python
#Japan, Philippines and Singapore should be part of APAC
customer[customer.TERRITORY == 'Japan'].groupby('COUNTRY').count()
```

```python
#Replace Japan with APAC
customer['TERRITORY'] = customer['TERRITORY'].str.replace('Japan', 'APAC')

customer.TERRITORY.unique()
```

```python
#Merge the data (customer & order)
customer_on_order = pd.merge(customer, order, on = ['ORDERNUMBER','PRODUCTCODE'], how='left')
```

```python
customer_on_order
```

```python
customer_on_order.info()
```

```python
#merge all the data (customer and order) & product
all_merged = pd.merge(customer_on_order,product, on='PRODUCTCODE', how = 'inner')
```
